```python
# Parameters
environment = "/obstacles_only/" # Note: all has obstacles + octaves
subEnvironment = "circle/"
subSubEnvironment = ""
trainOrTest = "train data/"
commonPath = os.getcwd() + environment + "heightmaps/" + subEnvironment + subSubEnvironment + trainOrTest
heightmap_path = commonPath + "inputs"
masks_path = commonPath + "masks"
labels_path = commonPath + "labels"

# Create directories
try:
    os.makedirs(masks_path)
except:
    print("Mask directory already exists...")

try:
    os.makedirs(labels_path)
except:
    print("Label directory already exists...")

img_list=os.listdir(heightmap_path)
filePath = commonPath + "completed_imgs.txt"
forMatlab_path = commonPath + "forMatlab.txt"



try:
    f = open(filePath, 'r')
except:
    f = open(filePath, 'w')
lines = f.readlines()
f.close()
```

```python

'''
Overview: 
Run through the img_list and starts a python macro script to run coppeliasim and generate lidar scans through
matlab. Note, matlab creates a label and mask that will be used for training with a GAN

Notes:
1) Time delays added at some critical parts to ensure that script runs smoothly

Bugs:
1) After running 2+ days, somehow matlab code indexing got very slightly rearranged, causing matlab to not run
    - Have not been able to reproduce bug
    - Current working soln: Drag matlab terminal up such that code is not visible
'''
prevTime = time.time()
for i in range(len(img_list)):
    file_name = img_list[i]
    if file_name+ "\n" not in lines:
        importHeightmap(i)
        time.sleep(1)

        runCoppeliaSimulation()
        time.sleep(10)

    #     switchToMatlab()
    #     time.sleep(2)

        runMatlabCode()
        time.sleep(2)

    #     switchToCoppeliaSim()
    #     time.sleep(2)

        stopCoppeliaSimulation()
        time.sleep(2)

        deleteHeightmap()
        time.sleep(1)

        isDoneProperly(i)
        if ((i+1) % 250) == 0:
            print("Current terrain: %d/%d" % (i+1, len(img_list)))
            resetCoppeliaSim()
            currentTime = time.time()
            print("Duration: %0.3fs" % (currentTime-prevTime))
            prevTime = currentTime

print("Done")
```
